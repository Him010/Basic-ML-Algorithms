```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
iris = datasets.load_iris()
x_train, x_test, y_train, y_test= train_test_split(iris.data, iris.target, random_state=1)
features= [1,2,3,4]
features= np.array(features)
```

```python
def print_leafNode(y,level):
    
    if(len(list(set(y)))==1):
        count_value= list(set(y))[0]
        count= np.count_nonzero(y==count_value)
        print("Level ",level)
        print("Count of ",count_value," ",count)
        print("Current Entropy  is = 0.0")
        print("Reached leaf Node")
        print()
 
    else:
        length= len(list(set(y)))
        
#         final_count_value=0
#         final_count=0
        
        print("Level ",level)
        for i in range(length):
            
            count_value= list(set(y))[i]
            count= np.count_nonzero(y==count_value)
            
#             if(count>final_count):
#                 final_count_value=count_value
#                 final_count=count
            print("Count of ",count_value," ",count)
        print("Current Entropy  is = 0.0")
        print("Reached leaf Node")
        print()
```

```python
def print_node(count0,count1,count2,feature_index, gainRatio ,level, features):
 
    print("Level ",level)
    print("Count of 0 = ",count0)
    print("Count of 1 = ",count1)
    print("Count of 2 = ",count2)
    print("Current Entropy  is = ",information(count0,count1,count2))
    print("Splitting on feature",features[feature_index],"with gain ratio",gainRatio)
    print()
    
```

```python
def split(X , Y , splitting_value , splitting_feature) :
    X_left  = [] 
    X_right = []
    Y_left = [] 
    Y_right = [] 
    
    for i in range(len(X)) :
        if X[i][splitting_feature] < splitting_value :
            X_left.append(X[i])
            Y_left.append(Y[i])
        else :
            X_right.append(X[i])
            Y_right.append(Y[i])
            
    X_left = np.array(X_left) 
    X_right = np.array(X_right)
    Y_left = np.array(Y_left) 
    Y_right = np.array(Y_right)
            
    X_left = np.delete(X_left , splitting_feature , axis = 1)
    X_right = np.delete(X_right , splitting_feature , axis = 1)
    
    return X_left , Y_left,  X_right , Y_right
```

```python
def information(count0,count1,count2):
    
    totalCount= count0+count1+count2
    
    a= count0/totalCount
    b= count1/totalCount
    c= count2/totalCount
    
    if(a==0):
        a=1
    if(b==0):
        b=1
    if(c==0):
        c=1
    
    info= -((a)*(np.log10(a))+ (b)*(np.log10(b))+ (c)*(np.log10(c)))
    
    return (info)
```

```python
def information_gain(count0InLeft, count1InLeft, count2InLeft, count0InRight, count1InRight, count2InRight):
    
    parentInfo =information(count0InLeft+count0InRight , count1InLeft+count1InRight, count2InLeft+count2InRight)
    
    leftChildInfo = information(count0InLeft, count1InLeft, count2InLeft)
    rightChildInfo = information(count0InRight, count1InRight, count2InRight)
    
    leftCount= count0InLeft + count1InLeft + count2InLeft
    rightCount= count0InRight + count1InRight + count2InRight
    
    totalCount= leftCount+ rightCount
    
    childInfo= (leftCount/totalCount)*(leftChildInfo) + (rightCount/totalCount)*(rightChildInfo)
    
    return (parentInfo - childInfo)
```

```python
def split_inforamtion(leftCount, rightCount):
    
    totalCount= leftCount+ rightCount
            
    a= leftCount/totalCount
    b= rightCount/totalCount
    
    if(a==0):
        a=1
    if(b==0):
        b=1
    
    info=  -((a)*(np.log10(a)) + (b)*(np.log10(b)))
    
    return (info)   
```

```python
def gain_ratio(x,y,feature_value):
    
    count0InLeft=0
    count1InLeft=0
    count2InLeft=0
    count0InRight=0
    count1InRight=0
    count2InRight=0
    
    for i in range(len(x)):
        
        if(x[i]<feature_value):
            if(y[i]==0):
                count0InLeft +=1
            elif(y[i]==1):
                count1InLeft +=1
            else:
                count2InLeft +=1
        else:
            if(y[i]==0):
                count0InRight+=1
            elif(y[i]==1):
                count1InRight+=1
            else:
                count2InRight+=1
                
    informationGain= information_gain(count0InLeft, count1InLeft, count2InLeft, count0InRight, count1InRight, count2InRight)
    splitInformation= split_inforamtion(count0InLeft + count1InLeft + count2InLeft, count0InRight + count1InRight + count2InRight)
    
    gainRatio= informationGain/splitInformation
    
    return gainRatio
```

```python
def max_gain_for_feature(x,y):
    
    continuousValuesOfFeature= list(set(x))               # list of values of one feature
    
    maxGainRatio=-1
    feature_check_value=-1
    for i in range(len(continuousValuesOfFeature)-1):
        
        feature_value= (continuousValuesOfFeature[i]+continuousValuesOfFeature[i+1])/2
        
        gainRatio= gain_ratio(x,y,feature_value)
        
        if(gainRatio> maxGainRatio):
            maxGainRatio=gainRatio
            feature_check_value=feature_value
    
    return maxGainRatio,feature_check_value
```

```python
def decision_tree_node(x,y,features,level):
    
    # base condition 
    # if no features are left
    if(len(features)==0):
        #print the details of node..
        print_leafNode(y,level)
        return
        
    # if pure node is achived    
    elif(len(list(set(y)))==1):
        #print the details of node..
        print_leafNode(y,level)
        return
        
    maxGainRatio=-1
    feature_index=-1
    feature_check_value=-1
    for i in range(len(features)):
        
        gainRatio,check_value =max_gain_for_feature(x[:,i],y)
        
#         print(features[i])
#         print(gainRatio)
#         print(check_value)
        
        if(gainRatio>maxGainRatio):
            maxGainRatio=gainRatio
            feature_index=i
            feature_check_value=check_value
        

    count0= np.count_nonzero(y==0)
    count1= np.count_nonzero(y==1)
    count2= np.count_nonzero(y==2)
    
    print_node(count0,count1,count2, feature_index , maxGainRatio,level, features)
    
    left_X,left_Y,right_X,right_Y= split(x,y,feature_check_value,feature_index)
    
    features= np.delete(features,feature_index)
    
    decision_tree_node(left_X,left_Y,features,level+1)
    decision_tree_node(right_X,right_Y,features,level+1)
```
