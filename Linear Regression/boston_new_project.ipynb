```python
def stg(data,lr,m):
    m_slope=np.zeros(len(m))
    M=len(data)
    for i in range(M):
        x=data[i,:-1]
        y=data[i,-1]
        for j in range(len(m)):
            m_slope[j]+=(-2/M)*(y-(x*m).sum())*x[j]
    return m-lr*m_slope
```

```python
def gd(data,lr,it):
    m=np.zeros(data.shape[1]-1)
    for i in range(it):
        m=stg(data,lr,m)
        print("cost",i+1,cost(data,m))
    return m
```

```python
def run(data):
    it=1000
    lr=0.0999999
    m=gd(data,lr,it)
    return m
```

```python
def cost(data,m):
    total_cost=0
    M=len(data)
    for i in range(M):
        x=data[i,:-1]
        y=data[i,-1]
        total_cost+=(1/M)*((y-(m*x).sum())**2)
    return total_cost
```

```python
import numpy as np
import pandas as pd
train_data=np.loadtxt("boston_train.csv",delimiter=',')
test_data=np.loadtxt("boston_test.csv",delimiter=',')
print(train_data.shape,test_data.shape)

x_train=train_data[:,:-1]
y_train=train_data[:,-1]
print(y_train.dtype)
ones=np.ones(len(train_data))
x_train=np.append(x_train,ones.reshape(-1,1),axis=1)
train_data=np.append(x_train,y_train.reshape(-1,1),axis=1)

x_test=test_data
ones=np.ones(len(test_data))
x_test=np.append(x_test,ones.reshape(-1,1),axis=1)
test_data = x_test
#test_data=np.append(x_test,y_test.reshape(-1,1),axis=1) 
# THERE IS NO Y_TEST IN TEST.CSV
print(train_data.shape,test_data.shape)
m=run(train_data)
print(m)
```

```python
def predict(x,m):
    y = []
    for i in range(len(x)):
        y.append((x[i]*m).sum())
    return y
```
