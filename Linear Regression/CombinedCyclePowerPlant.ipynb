```python
import numpy as np
```

```python
def step_gradient(points,learning_rate,m_array):
    M = len(points)
    features = len(points[0])
    m_slope_array = np.zeros(features-1)
    new_m_array = np.zeros(features-1)
    for i in range(M): 
        x = points[i,: -1]
        y = points[i,-1]
        for j in range(features-1):
#             m_slope_array[i] += (-2/M) * ((y - sum1[j]) * x)
            m_slope_array[j] += (-2/M) * ((y - (x * m_array).sum()) * x[j])
    new_m_array = m_array - learning_rate * m_slope_array
    return new_m_array
```

```python
def gd(points,learning_rate,num_iterations):
    features = len(points[0])
    m_array = np.zeros(features-1)
    for i in range(num_iterations):
        m_array = step_gradient(points,learning_rate,m_array)
        print(i, " Cost: ", cost(points,m_array))
    return m_array
```

```python
def cost(points,m_array):
    total_cost = 0
    M = len(points)
    for j in range(M):
        x = points[j, :-1]
        y = points[j,-1]
        total_cost += (1/M) * ((y - (m_array * x).sum()) ** 2)
    return total_cost
```

```python
def predict(X_test,m_array):
#     M = len(points)
#     features = len(points[0])
    Y_pred = []
    for i in range(len(X_test)):
        Y_pred.append((X_test[i]*m_array).sum())     # taking one row of the training data and usig m_array. multiplying them 
                                                     # will give us a row equal to size of the one row of training data
                                                     # and summing them will give us m1*x1+m2*x2+m3*x3 ....
    np.savetxt("boston_NEW_pred2.csv",Y_pred)
    return Y_pred
```

```python
def score(Y_test,Y_pred):          # COD is Coefficient of Determination
    u = ((Y_test - Y_pred) ** 2).sum()
    v = ((Y_test - Y_test.mean()) ** 2).sum()
    score = 1 - u/v
    return score
```

```python
def run():
    data = np.loadtxt("ccpp_x_y_train.csv",delimiter = ",")
    test_data = np.loadtxt("ccpp_x_test.csv",delimiter = ",")
    num_iterations = 20000
    learning_rate = 0.000000955
    features = len(data[0])
    ones = np.ones(len(data))
    data = np.insert(data,len(data[0])-1,ones,axis=1)
    m_array = []
    m_array = gd(data,learning_rate,num_iterations)
    print()
    print(m_array)
    Y_pred = predict(data[: , 0 : features],m_array)#,data)
    y = data[ : , -1]
    finalScore = 0
    finalScore = score(y,Y_pred)
    print("The final score is : " , finalScore)
    ones = np.ones(len(test_data))
    test_data = np.insert(test_data,len(test_data[0]),ones,axis=1)
    y_final_pred = predict(test_data,m_array)#,test_data)
    print()
    print()
    print(y_final_pred)
    print()
    print()
```
