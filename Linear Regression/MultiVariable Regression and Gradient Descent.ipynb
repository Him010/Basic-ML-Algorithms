# Complex Boundaries

```python
from sklearn import datasets
```

```python
boston = datasets.load_boston()
X = boston.data
Y = boston.target
```

```python
import pandas as pd
df = pd.DataFrame(X)
df.columns = boston.feature_names
df["AGE_AGE"] = df.AGE ** 2     # can also write it as df["AGE_AGE"] = df["AGE"] ** 2

# X contains the previous data. so let's make a new X2 that contains the new data with the feature AGE_AGE
X2 = df.values
df.describe()
```

```python
from sklearn import model_selection
X_train,X_test,Y_train,Y_test = model_selection.train_test_split(X,Y, random_state = 0)
X2_train,X2_test,Y2_train,Y2_test = model_selection.train_test_split(X2,Y, random_state = 0)

# SO NOW IT'S GOING TO CREATE EXACTLY THE SAME RANDOM NUMBERS BECAUSE IT'S STARTING FROM THE SAME STATE, SO THE SAME ROWS WILL
# GO INTO WHICHEVER DIRECTION WE TAKE

# IT IS NECESSARY BECAUSE NOT PROVIDING A RANDOM STATE IS NOT A GOOD OPTION WHILE COMPARING AS THE TRAINING DATA OF ONE MIGHT
# BE THE TESTING DATA OF THE OTHER SO IT MIGHT NOT GIVE CORRECT COMPARING RESULTS

# WHAT SPLIT() DOES IT CALLS RANDOM NUMBERS AND THEN SPLIT THE DATA ACCORIDNG TO THAT IN TRAINING AND TESTING ACCORDINGLY
# WHAT RANDOM STATE DOES IS IT GENERATES THE SAME RANDOM NUMBERS FOR BOTH THE SPLITS SO NOW THE DATA GOING IN TRAINING OF ONE
# WOULD BE THE SAME AS THE DATA GOING INTO TRAINING OF OTHER AND SAME WITH THE TESTING DATA AS WELL
```

```python
from sklearn.linear_model import LinearRegression
alg1 = LinearRegression()
alg2 = LinearRegression()
```

```python
alg1.fit(X_train,Y_train)
alg2.fit(X2_train,Y2_train)
```

```python
Y_pred = alg1.predict(X_test)
Y2_pred = alg2.predict(X2_test)
```

```python
print("Score of the training data is " , alg1.score(X_train,Y_train))
print("Score of the testing data is ", alg1.score(X_test,Y_test))
```

```python
print("Score of the NEW training data is " , alg2.score(X2_train,Y2_train))
print("Score of the NEW testing data is ", alg2.score(X2_test,Y2_test))
```

# BELOW TWO COLUMNS ARE USED TO ADD DIFFERENT FEATURES USING FOR LOOPS

```python
for i in range(13):
    name = df.columns[i]
    df[i] = df[name] ** 2
df.describe()
```

```python
for i in range(13):
    for j in range(i+1,13):
        name1 = df.columns[i]
        name2 = df.columns[j]
        df[i,j] = df[name1] * df[name2]
df.describe()
```
