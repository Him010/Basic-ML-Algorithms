```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
np.random.seed(2342343)
```

```python
mean_vac1 = np.array([0,0,0])
cov_mat1 = np.array([[1,0,0],[0,1,0],[0,0,1]])
class1 = np.random.multivariate_normal(mean_vac1,cov_mat1,100) 
```

```python
mean_vac2 = np.array([1,1,1])
cov_mat2 = np.array([[1,0,0],[0,1,0],[0,0,1]])
class2 = np.random.multivariate_normal(mean_vac2,cov_mat2,100) 
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D, proj3d

fig = plt.figure(figsize =  (8,8))
ax = fig.add_subplot(111,projection = "3d")
ax.plot(class1[:,0],class1[:,1],class1[:,2],'o')
ax.plot(class2[:,0],class2[:,1],class2[:,2],'^')
plt.show()
```

```python
all_data = np.concatenate((class1,class2))
```

```python
pca = PCA(n_components = 2)
transformed_data = pca.fit_transform(all_data)
transformed_data
```

```python
pca.components_
```

```python
plt.scatter(transformed_data[0:100,0],transformed_data[0:100,1])
plt.scatter(transformed_data[100:200,0],transformed_data[100:200,1])
plt.show()
```

```python
X_approx = pca.inverse_transform(transformed_data)
fig = plt.figure(figsize =  (8,8))
ax = fig.add_subplot(111,projection = "3d")
#ax.plot(class1[:,0],class1[:,1],class1[:,2],'o')
ax.plot(X_approx[:,0],X_approx[:,1],X_approx[:,2],'^')
plt.show()
```

```python
all_data_T = all_data.T
cov = np.cov(all_data_T)
cov
```

```python
eig_val,eig_vectors = np.linalg.eig(cov)
eig_val,eig_vectors
```

```python
eig_val_vector_pair = []
for i in range(len(eig_val)):
    eig_vec = eig_vectors[:,i]
    eig_val_vector_pair.append((eig_val[i],eig_vec))
    
eig_val_vector_pair.sort(reverse = True)
eig_val_vector_pair
```
