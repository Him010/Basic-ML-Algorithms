```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3]
y = [2,4,6]
plt.scatter(x,y)             # to plot them as scattered points
plt.show()
```

```python
x = [1,2,3]
y = [2,4,6]
plt.plot(x,y)              # to plot them as a single line connecting them together
plt.scatter(x,y)
plt.show()
```

```python
x = [1,2,15]
y = [2,4,6]

x2 = [3,4,5]
y2 = [2,4,6]

plt.plot(x,y)
plt.plot(x2,y2, "y--")             # y i.e. yellow decides the color of the line joining points x2 and y2
                                   # -- decides we want the line to be dashed
```

```python
# plot function x^3
# x = [1,2,3]
# x**3                # NOT VALID as it is a list and we can't cube list by just doing this
# plt.plot(x,x**3)
# plt.show()

# to easily cube an array. use numpy array. so x**3 will work on that

import numpy as np
x = np.array([1,2,3,4])
y = x ** 3
plt.plot(x,y)
plt.show()
```

```python
x = np.arange(0,5,0.1)         # 0 is the starting point. 5 is the ending point. 0.1 is the increment
                               # so the array would look like x = [0,0.1,0.2,0.3,0.4.......4.9]
y = x ** 3
plt.plot(x,y)
plt.show()
```

```python
a = [3,4,5]      # it will take these three points as y
                 # x will be since three points so x = [0,1,2]
                 # if y was n points so x would be range(n) i.e. x = [0,1,2....n-1]
plt.plot(a)
plt.show()
```

```python
# Customising Graphs

x = np.arange(0,5,0.1)
y = x ** 3
y2 = x ** 2
# plt.plot(x,y,color = "blue", marker = "o")
plt.plot(x,y,color = "blue",linewidth = 5, label = "x^3")
plt.plot(x,y2,color = "red",linewidth = 5,label = "x^2")

plt.legend()            # if you want the label to appear showing x^3 and x^2 then you want to call this function

# plt.axis([0,10,0,100])
plt.grid()
plt.ylabel("y")
plt.xlabel("x")
plt.title("Matplotlib Demo")
plt.text(2,80,"Text",fontsize = 20)
plt.show()
```

# Pie Graph

```python
sizes = [3,4,6,2]
label = ["A","B","C","D"]
color = ["blue","purple","red","pink"]
plt.title("Split Among Classes",fontsize = 20)
explode = [0.1,0,0,0]

# plt.pie(sizes,colors = color,labels = label,autopct="%.1f")

# autopct will automatically whow the percentage by using "%.1f" or you can pass a function in autopct and it will return a
# string and you can print that too


plt.pie(sizes,colors = color,labels = label,autopct="%.1f%%",explode = explode,counterclock = False,startangle = 100)
# if you want to add "%" in the end, do autopct = "%.1f%%"
# counterclock = False will make the pie graph clockwise

# startangle = 100 will make the graph start from 100 degree. initially it started from 0 degree but with startangle = 100 degree
# it would start from a 100 degree


plt.axis("equal")        # to make the axis equal i.e. make the pie exactly circular. usually in this python verison 
                         # they are equal but if not then use this command to make them equal
plt.show()
```
