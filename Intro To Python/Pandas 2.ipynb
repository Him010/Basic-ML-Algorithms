## Handling NaN

```python
# Manipulating Data in Data-Frame 2
import pandas as pd
import numpy as np
iris = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data")
```

```python
df = iris.copy()
df.columns = ['sl','sw','pl','pw','flower_type']
```

```python
df["diff_pl_pw"] = df["pl"] - df["pw"]    # adds a new column which is the difference of pl and pw
df["abc"] = 1                             # makes the column of all ones
df.head()
```

```python
df.iloc[2:4,1:3] = 0
df.head()
```

```python
df.iloc[2:4, 1:3] = np.nan     # to make the entries NaN
df.head()
df.describe()
```

```python
df.dropna(inplace = True)
df.head()
```

```python
df.reset_index(drop = True, inplace = True)
df.head()
```

```python
df.iloc[2:4, 1:3] = np.nan
df.head()
```

```python
df.sw.fillna(df.sw.mean(),inplace = True)
df.head()
```

```python
df.pl.fillna(df.pl.mean(),inplace = True)
df.head()
```

## Handling Strings in Data

```python
# we generally want our column data to be numeric. In most cases we'd want to change the data to a numerical data
# String based Data
df["Gender"] = "Female"
df.iloc[0:10, 7] = "Male"
df.head()

def f(s):
    if s == "Male":
        return 0
    else:
        return 1
    
df["sex"] = df.Gender.apply(f)
del df["Gender"]
df.head()
```
