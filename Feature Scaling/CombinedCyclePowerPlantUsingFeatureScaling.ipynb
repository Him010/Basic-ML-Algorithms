```python
import numpy as np
from sklearn import preprocessing
```

```python
def step_gradient(pointsX,pointsY,learning_rate,m_array):
    M = len(pointsX)
    features = len(pointsX[0])
    m_slope_array = np.zeros(features)
    new_m_array = np.zeros(features)
    for i in range(M): 
        x = pointsX[i,:]
        y = pointsY[i]
        for j in range(features):
#             m_slope_array[i] += (-2/M) * ((y - sum1[j]) * x)
            m_slope_array[j] += (-2/M) * ((y - (x * m_array).sum()) * x[j])
    new_m_array = m_array - learning_rate * m_slope_array
    return new_m_array
```

```python
def gd(pointsX,pointsY,learning_rate,num_iterations):
    features = len(pointsX[0])
    m_array = np.zeros(features)
    for i in range(num_iterations):
        m_array = step_gradient(pointsX,pointsY,learning_rate,m_array)
        print(i, " Cost: ", cost(pointsX,pointsY,m_array))
    return m_array
```

```python
def cost(pointsX,pointsY,m_array):
    total_cost = 0
    M = len(pointsX)
    for j in range(M):
        x = pointsX[j,:]
        y = pointsY[j]
        total_cost += (1/M) * ((y - (m_array * x).sum()) ** 2)
    return total_cost
```

```python
def predict(X_test,m_array):
#     M = len(points)
#     features = len(points[0])
    Y_pred = []
    for i in range(len(X_test)):
        Y_pred.append((X_test[i]*m_array).sum())     # taking one row of the training data and usig m_array. multiplying them 
                                                     # will give us a row equal to size of the one row of training data
                                                     # and summing them will give us m1*x1+m2*x2+m3*x3 ....
    np.savetxt("CombinedPowerPlantUsingFeatureScaling.csv",Y_pred)
    return Y_pred
```

```python
def score(Y_test,Y_pred):          # COD is Coefficient of Determination
    u = ((Y_test - Y_pred) ** 2).sum()
    v = ((Y_test - Y_test.mean()) ** 2).sum()
    score = 1 - u/v
    return score
```

```python
def run():
    train_data = np.loadtxt("ccpp_x_y_train.csv",delimiter = ",")
    normal_test_data = np.loadtxt("ccpp_x_test.csv",delimiter = ",")
    normal_train_data_X = train_data[:,:-1]
    normal_train_data_Y = train_data[:,-1]
    
    # Feature Scaling
    scaler = preprocessing.StandardScaler()
    scaler.fit(normal_train_data_X)
    test_data = scaler.transform(normal_test_data)
    data = scaler.transform(normal_train_data_X)
    
    num_iterations = 1000
    learning_rate = 0.403
#     features = len(data[0])
    ones = np.ones(len(data))
    data = np.insert(data,len(data[0]),ones,axis=1)
    features = len(data[0])
    m_array = []
    m_array = gd(data,normal_train_data_Y,learning_rate,num_iterations)
    print()
    print(m_array)
    Y_pred = predict(data,m_array)#,data)
    y = normal_train_data_Y
    finalScore = 0
    finalScore = score(y,Y_pred)
    print("The final score is : " , finalScore)
    ones = np.ones(len(test_data))
    test_data = np.insert(test_data,len(test_data[0]),ones,axis=1)
    y_final_pred = predict(test_data,m_array)#,test_data)
    print()
    print()
    print(y_final_pred)
    print()
    print()
```
